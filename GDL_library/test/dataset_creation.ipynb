```python
from gdl.data.datasets import RepresentationDataset, RepresentationInMemoryDataset
from gdl.data.featurization import ResidueFeaturizer
from gdl.data.representations import RepresentationBuilder

import pandas as pd
```

# Data

Here we create the dictionary for all enzimes in the raw data folder

```python
raw_data_dir = "../data/DeepET/pdb_temp.csv"

temp_df = pd.read_csv(raw_data_dir, index_col=0)
temp_df["pdb_id"] = temp_df["pdb_id"] + "_cleaned_centered.pdb"
temp_dict = temp_df.set_index('pdb_id').to_dict()["topt"]
temp_dict
```

```python
temp_dict["1A0E_cleaned_centered.pdb"]
```

# Dataset Creation

To create a dataset we need:

```python
featurizer = ResidueFeaturizer()
builder = RepresentationBuilder(representation="graph", featurizer=featurizer)
```

```python
raw_data_dir = "../data/DeepET/pdb_clean"
root = "../data/DeepET/rep_dataset"
dataset_name = "example_dataset"

dataset = RepresentationDataset(raw_data_dir=raw_data_dir, root=root, dataset_name=dataset_name,
                                 builder=builder, label_dict=temp_dict)
```

```python
dataset[0]
```

```python
root = "../data/DeepET/rep_im_dataset"
im_dataset = RepresentationInMemoryDataset(raw_data_dir=raw_data_dir, root=root, dataset_name=dataset_name,
                                 builder=builder, label_dict=temp_dict)
```
